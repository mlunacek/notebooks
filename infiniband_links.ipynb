#Infiniband links

```python
import pandas as pd
import numpy as np
import glob
```

```python
files = glob.glob('/home/molu8455/projects/infiniband/learning/data/link/*.csv')

frames = []
for f in files:
    tmp = pd.read_csv(f)
    # check tmp...
    frames.append(tmp)
    
df = pd.concat(frames)
df.columns = ['id', 'same','result','device'] # better names

print len(df)
```

```python
df.head()
```

```python
df = df.dropna()
df = df[df.result != 'None']
df.result = df.result.astype(np.float)
print len(df)
```

```python
jobs = df.groupby('id').mean()
jobs.head()
```

##Failures and links

```python
fail_pv = pv[jobs.result > threshold*1000] 
pass_pv = pv[jobs.result < threshold*1000] 
```

```python
fail_sum = fail_pv.sum(axis=0)
pass_sum = pass_pv.sum(axis=0)
diff = fail_sum - pass_sum
percent = (fail_sum)/(fail_sum + pass_sum)
```

```python
index = np.argsort(pass_sum.values)
plot(fail_sum[index], color="red")
plot(pass_sum[index], color="green", linewidth=3)
show()
```

```python
plot(fail_sum[index], pass_sum[index], 'o', alpha=0.25)
plot(range(900), range(900))
xlim(0,900) # <- Fix the failures axis
xlabel('failures')
ylabel('pass')
title('total link usage')
show()
```

```python
boxplot([fail_sum[index], pass_sum[index]], widths=0.65)
ylim(0,500)
show()
```

This is not significant.

```python
index = np.argsort(fail_sum.values)
plot(pass_sum[index], color="green")
plot(fail_sum[index], color="red", linewidth=3)
show()
```

```python
index = argsort(percent.values)
plot(percent[index], color="red")
plot(1-percent[index], color="green")
show()
```

How many links only occur in failed jobs?
