## Best solution for each trial

```python
from functools import partial

def read_df(trial, files):
    
    df = pd.read_csv(files[trial], header=None)
    df = df.dropna(axis=0)
    df['trial'] = trial
    return df

part_read_df = partial(read_df, files=files)

df = pd.concat(map(part_read_df, range(len(files))))
```

```python
def best_trial(df):
    best = df.sort(0, ascending=True).head(1).index
    return df.ix[best]

best = df.groupby('trial').apply(best_trial)
```

```python
plt.hist(best.ix[:,0].values, bins=10)
plt.xlabel('objective function value (e.g. fit)')
plt.show()
```

**NOTE**: Most of the trials found very competative solutions.
