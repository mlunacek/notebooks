# Sensitivity Analysis and the Hessian

```python
import smash
from ipywidgets import StaticInteract, RangeWidget

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import os
import glob
import numpy as np

d_path = '/home/molu8455/projects/rengers/optimization_cma_es'
files = glob.glob(os.path.join(d_path, 'xxx*'))
files
```

```python
def plot(gen):
    C = C_array[gen]
    fig, ax = plt.subplots(1,2, figsize=(14,8))
    ax[0].imshow(np.linalg.inv(C), cmap=plt.cm.Greys)
    height = list(np.diag(np.linalg.inv(C)))
    index = range(len(height))
    ax[1].bar(index, height, 0.5, color="grey")
    return smash.format(fig)

def get_c_array(index):
    f = files[index]
    with open(f,'r') as infile:
        data = infile.read()

    C_array = []
    for line in data.split('COV')[1:]:
        tmp = line.split('FIT')[0]
        tmp = tmp.strip('\n').replace('\n','').replace('[','').replace(']','')
        str_array = tmp.split()
        try:
            C = np.array(map(lambda x: float(x), str_array)).reshape(11,11)
            #eigen_array.append(np.linalg.eig(C)[0])
            C_array.append(C)
        except ValueError:
            pass    
    return C_array

def eigen_values(index):
    f = files[index]
    with open(f,'r') as infile:
        data = infile.read()
        
    eigen_array = []
    for line in data.split('COV')[1:]:
        tmp = line.split('FIT')[0]
        tmp = tmp.strip('\n').replace('\n','').replace('[','').replace(']','')
        str_array = tmp.split()
        try:
            C = np.array(map(lambda x: float(x), str_array)).reshape(11,11)
            eigen_array.append(np.linalg.eig(C)[0])
            #eigen_array.append(np.diag(C))
        except ValueError:
            pass
        
    return np.array(eigen_array).reshape(len(eigen_array),11)

def diag_values(index):
    f = files[index]
    with open(f,'r') as infile:
        data = infile.read()
        
    eigen_array = []
    for line in data.split('COV')[1:]:
        tmp = line.split('FIT')[0]
        tmp = tmp.strip('\n').replace('\n','').replace('[','').replace(']','')
        str_array = tmp.split()
        try:
            C = np.array(map(lambda x: float(x), str_array)).reshape(11,11)
            #eigen_array.append(np.linalg.eig(C)[0])
            eigen_array.append(np.diag(C))
        except ValueError:
            pass
        
    return np.array(eigen_array).reshape(len(eigen_array),11)

def final_cov(index):
    return get_c_array(index)[-1]

def plot_eig(index):
    fig, ax = plt.subplots(1,2,figsize=(10,6))
    eig_array = eigen_values(index)    
    for c in range(eig_array.shape[1]):
        ax[0].plot(eig_array[:,c])
    ax[0].set_title('{0} {1}'.format(index,os.path.basename(files[index])))
    
    C = final_cov(index)
    ax[1].imshow(np.linalg.inv(C), cmap=plt.cm.Greys)
    return smash.format(fig)

def plot_diag(index):
    fig, ax = plt.subplots(1,2,figsize=(10,6))
    eig_array = diag_values(index)    
    for c in range(eig_array.shape[1]):
        ax[0].plot(eig_array[:,c])
    ax[0].set_title('{0} {1}'.format(index,os.path.basename(files[index])))
    
    C = final_cov(index)
    ax[1].imshow(np.linalg.inv(C), cmap=plt.cm.Greys)
#     height = list(np.diag(np.linalg.inv(C)))
#     index = range(len(height))
#     ax[2].bar(index, height, 0.5, color="grey")
    
    return smash.format(fig)
```

## Changing eigenvalues

```python
StaticInteract(plot_eig, index=RangeWidget(0, len(files)-1))
```
